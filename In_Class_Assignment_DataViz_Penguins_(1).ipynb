## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.show()
```

**Your interpretation (2–3 sentences):**

_Write your observations here._

Adelie is represented by blue and appears to have the shortest bill length among the species, but has a similar bill depth to the chinstrap. Chinstrap is represented by orange and has a similar bill length to the Gentoo, but has a higher bill depth. The Gentoo has the smallest bill depth among the species.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

penguins.hist(column='bill_length_mm', bins=10)
plt.show()
```

```python
# TODO: Histogram with 30 bins

penguins.hist(column='bill_length_mm', bins=30)
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._

The first histogram is more condensed and gives a good summary of the information and its shape. However, the 30 bin histogram provides a more detailed representation and allows us to see the dip in the middle, followed by a large spike after the 45 mark.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.show()
```

**Your summary (1–2 sentences):**
